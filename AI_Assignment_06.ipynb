```python
import numpy as np

def initialize_weights(n_features):
    weights = np.zeros(n_features)
    bias = 0.0
    return weights, bias

def step_activation(z):
    return 1 if z >= 0 else 0

def predict(X, weights, bias):
    predictions = []
    for x in X:
        z = np.dot(x, weights) + bias
        predictions.append(step_activation(z))
    return np.array(predictions)

def train_perceptron(X, y, learning_rate=0.1, epochs=20):
    weights, bias = initialize_weights(X.shape[1])

    for epoch in range(epochs):
        errors = 0
        for xi, yi in zip(X, y):
            z = np.dot(xi, weights) + bias
            y_pred = step_activation(z)
            error = yi - y_pred
            weights += learning_rate * error * xi
            bias += learning_rate * error
            errors += int(error != 0)

        accuracy = (len(y) - errors) / len(y) * 100
        print(f"Epoch {epoch+1:02d} | Errors: {errors} | Accuracy: {accuracy:.1f}%")

    return weights, bias

def evaluate(X, y, weights, bias):
    preds = predict(X, weights, bias)
    accuracy = np.mean(preds == y) * 100
    print(f"\nFinal Test Accuracy: {accuracy:.1f}%")
    for i, (xi, yi, pi) in enumerate(zip(X, y, preds)):
        print(f"  Sample {i+1}: Input={xi} | True={yi} | Pred={pi} | {'✓' if yi==pi else '✗'}")

# AND gate dataset
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])
y = np.array([0, 0, 0, 1])

print("=== Single-Layer Perceptron Training (AND Gate) ===\n")
weights, bias = train_perceptron(X, y, learning_rate=0.1, epochs=20)

print(f"\nLearned Weights: {weights} | Bias: {bias:.2f}")
print("\n=== Evaluation ===")
evaluate(X, y, weights, bias)
```
